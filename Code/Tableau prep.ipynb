```python
latdict = {1:41.995,2:41.786}
longdict = {1:-87.933,2:-87.752}
```

```python
weather_data['Trap']=map(lambda x: 'Chicago O\'Hare International Airport' if x==1 else 'Chicago Midway International Airport',weather_data.Station)
weather_data['Latitude'] = map(lambda x: latdict[x], weather_data.Station)
weather_data['Longitude'] = map(lambda x: longdict[x], weather_data.Station)
```

```python
weather_data.Trap.value_counts()
```

```python
all_data = tableau.append(weather_data)
all_data.dtypes
all_data.Tavg = all_data.Tavg.astype('float64', errors='ignore')
```
